```python
import json
import requests as req
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from citipy import citipy
import random 
import decimal
```

```python
lat_lngs=[]
cities = []

#create a set of random lat and lng combo

lats= np.random.uniform(low=-90.00, high=90.00, size= 1500)
lngs = np.random.uniform(low=-180.00,high=180.00, size= 1500)

lat_lngs = zip(lats,lngs)


for lat_lng in lat_lngs:
    city= citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    if city not in cities:
        cities.append(city)
len(cities)
```

```python
cities = pd.DataFrame(data =cities)
cities['Cities'] = cities
cities = pd.DataFrame(data= cities['Cities'])
```

```python
cities_sample = cities.sample(n=400)
cities_sample_test= cities_sample.reset_index(drop = True)
```

```python
cities_sample_test['Country'] = ''
cities_sample_test['Latitude'] = ''
cities_sample_test['Longitude'] = ''
cities_sample_test['Temperature'] = ''
cities_sample_test['Cloudiness'] = ''
cities_sample_test['Humidity'] = ''
cities_sample_test['Windspeed (mph)'] = ''
```

```python
cities_sample_test1 = cities_sample_test
```

```python
# set up api 
api_key = 'e49020bc745592e6faec88f330db1b5b'
cities_zero = cities_sample_test['Cities'][1]
target_url = "http://api.openweathermap.org/data/2.5/forecast?q=%s&units=IMPERIAL&mode=json&APPID=%s" %  (str(cities_zero),api_key)
weather_data = req.get(target_url).json()
```

```python
counter = -1
set_counter = 0
city_list = []
for index, row in cities_sample_test1.iterrows():
        test_url = "http://api.openweathermap.org/data/2.5/forecast?q=%s&units=IMPERIAL&mode=json&APPID=%s" %  (row['Cities'].replace(" ",""),api_key)
        results_test = req.get(test_url).json()
        
        try:
            cities_sample_test1.set_value(index, "Latitude",results_test['city']['coord']['lat'])
            cities_sample_test1.set_value(index, "Country",results_test['city']['country'])
            cities_sample_test1.set_value(index, "Longitude",results_test['city']['coord']['lon'])
            cities_sample_test1.set_value(index, "Temperature",results_test['list'][0]['main']['temp_max'])
            cities_sample_test1.set_value(index, "Humidity",results_test['list'][0]['main']['humidity'])
            cities_sample_test1.set_value(index, "Windspeed (mph)",results_test['list'][1]['wind']['speed'])
            cities_sample_test1.set_value(index, "Cloudiness",results_test['list'][0]['clouds']['all'])
            counter += 1
            print(test_url)
            print('Processing Record '+ str(counter) + ' | ' + (row['Cities']))
            city_list.append(row['Cities'])
        except KeyError:
            print("City Not Found Skipping")
           

```

```python

refined_city_df = cities_sample_test1[cities_sample_test1['Cities'].isin(city_list)]
```

```python
#make scatter plot for Temperature (F) vs. Latitude
plt.figure(figsize=(7,5))
plt.scatter(refined_city_df['Latitude'],refined_city_df['Temperature'],marker= 'o', alpha = .75)

plt.title('Temperature (F) vs. Latitude on %s' %(date)) 
plt.xlabel('Latitude')
plt.ylabel('Temperature in (F)')
plt.xlim([-80,85])
plt.ylim([-10,110])
plt.tight_layout()
plt.grid()
plt.show()
```
